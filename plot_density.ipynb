```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/home/phu/Documents/dimethyl_sulfoxide_pkin_den_form.csv")
```

```python
plt.scatter(df.packing_index,df.Density, s=5, c="blue")
plt.xlabel("Packing index")
plt.ylabel("Density")
plt.title("Packing index vs Density")
plt.grid()
plt.gcf().set_size_inches(10,10)
plt.savefig("/home/phu/Documents/packingvsdensity.png")
```

```python
df = df.rename(index=str, columns={"Unnamed: 0": "Index"})
```

```python
df.columns
```

```python
df2 = pd.DataFrame(columns=['Index', 'Name', 'Formula', 'packing_index', 'Density'])
```

```python
df2 = pd.DataFrame(columns=['Index', 'Name', 'Formula', 'packing_index', 'Density'])
for i,row in df.iterrows():
    n = row["Formula"]
    if "Na" in n:
        df2 = df2.append(row)
```

```python
objects = tuple(df2.Formula)
y = np.arange(len(objects))
plt.bar( y,df2.packing_index, align='center', alpha=0.5)
plt.gcf().set_size_inches(10,10)
plt.title("Na: Packing Index")
plt.ylabel("Packing index")
plt.savefig("/home/phu/Documents/Na_packingindex.png")
```
